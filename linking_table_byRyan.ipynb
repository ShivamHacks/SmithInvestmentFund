```python
import pandas as pd
import numpy as np
import datetime
```

```python
compustat = pd.read_csv('C:/Users/ryanm_fa0xwnc/PycharmProjects/master/Python-Projects/SIF/Data/Compustat19612018.csv')
CRSP = pd.read_csv('C:/Users/ryanm_fa0xwnc/PycharmProjects/master/Python-Projects/SIF/Data/CRSP2007_2018.csv')
linking_table = pd.read_csv('C:/Users/ryanm_fa0xwnc/PycharmProjects/master/Python-Projects/SIF/Data/CompustatMergedDatabase.csv')
```

```python
# convert integer dates in format YYYYMMDD to datetime objects in LINKDT and LINKENDDT columns.
# coerce errors because of 'E' values and then fill properly with date.today()
linking_table.LINKDT = pd.to_datetime(linking_table.LINKDT.astype(str), format='%Y%m%d', errors='coerce').fillna(datetime.date.today())
linking_table.LINKENDDT = pd.to_datetime(linking_table.LINKENDDT.astype(str), format='%Y%m%d', errors='coerce').fillna(datetime.date.today())
```

```python
linking_table.head()
```

```python
# Same as linking table, except error handling is not needed
compustat.datadate = pd.to_datetime(compustat.datadate.astype(str), format='%m/%d/%Y')
```

```python
compustat.head()
```

```python
# link compustat and linking_table on gvkey of both
compustat_linking = compustat.merge(linking_table, how='outer', left_on=['gvkey'], right_on=['gvkey'], validate='many_to_many')
```

```python
compustat_linking.head()
```

```python
# Filter out rows that are not valid, which are those where the date does not fall between the LINKDT and LINKENDDT
compustat_linking_filtered = compustat_linking[(compustat_linking.datadate >= compustat_linking.LINKDT) & (compustat_linking.datadate <= compustat_linking.LINKENDDT)]
compustat_linking_filtered.head()
```

```python
# merge compustat_linking_filtered and CRSP using LPERMCO from linking table and PERMNO fron CRSP to compustat_crsp
compustat_crsp = compustat_linking_filtered.merge(CRSP, how='outer', left_on=['LPERMCO'], right_on=['PERMNO'])
```
